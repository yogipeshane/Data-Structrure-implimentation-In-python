```python
class Node:
    def __init__(self, data, nxt=None):
        self.data = data
        self.nxt = nxt
```

```python
class LikedList :
    def __init__(self):
        self.head = None
        self.length = 0
        
    def insertNext(self, data):
        if self.head == None:
            node = Node(data)
            self.head = node
            self.length += 1
        else:
            itr = self.head
            
            while itr.nxt != None:
                itr  = itr.nxt
               
            node = Node(data)
            itr.nxt = node 
            self.length += 1
            
    def insertBetween(self, index, data):
        itr = self.head
        if itr == None or self.length < index:
            print("Error : Index out of bound!")
        else :
            llIndex = 0   

            while itr.nxt != None:                   
                if llIndex == index-1:
                    oldNode = itr.nxt
                    newNode = Node(data)
                    newNode.nxt = oldNode
                    itr.nxt = newNode
                    self.length+=1
                itr = itr.nxt                   
                llIndex +=1 
    
    def insertEnd(self, data):        
        lastNode = self.head
        if lastNode is None:            
            node= Node(data)
            self.head = node
            self.length += 1
        else:
            while lastNode.nxt != None:
                lastNode = lastNode.nxt
            node = Node(data)
            lastNode.nxt = node   
            self.length += 1    
            
    def extendList(self,elements):
        if elements == []:  
            print("Error: Given list is empty!")
        else:
            for item in elements:
                self.insertEnd(item)
                
    def get_elementAtIndex(self, index):
        if index > self.length:
            print("Error: Index out of Bound !")
        else:
            itr = self.head
            eleCount  = 0
            while itr.nxt != None:                
                if eleCount == index:
                    print("Element at {} Index is {}".format(index, itr.data))
                    break;
                itr = itr.nxt
                eleCount+=1
            
    def printLL(self):
        itr = self.head
        if itr != None:     
            while itr != None:
                print(itr.data)
                itr = itr.nxt
        else :
            print("Linked List not available")    
    
    def removeElement(self,data):
        if self.length == 0:
            print("Linked list is empty, We can not remove element from Linked List!")
        else:
            headNode = self.head
            if headNode.data == data:
                del headNode
            else:
                itr = headNode.nxt
                prevNode = headNode
                while itr.nxt != None:                    
                    if itr.data == data:
                        nxtNode = itr.nxt
                        prevNode.nxt = nxtNode
                        del itr
                        self.length -= 1
                        break
                    itr= itr.nxt
                    prevNode = prevNode.nxt
                    
    
```

```python
ll = LikedList()

ll.insertNext(4)
ll.insertNext(5)
ll.insertNext(6)
ll.insertBetween(1,7)
ll.printLL()
print("length :",ll.length)
ll.get_elementAtIndex(2)
ll.removeElement(5)
ll.printLL()
print("length :",ll.length)
ll.insertBetween(2,11)
ll.printLL()
print("length :",ll.length)
```
